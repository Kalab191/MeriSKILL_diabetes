```python
def mean_imputer(df, col): 
    #replaces the zeros(0) with the mean value
    mean = df[col].mean()
    df[col].replace(0, mean, inplace=True)

def median_imputer(df, col):
    # replaces the zeros(0) with the median value
    median = df[col].median()
    df[col].replace(0, median, inplace=True)
```

```python
mean_imputer(df, 'Glucose')
mean_imputer(df, 'SkinThickness')
median_imputer(df, 'BMI')
median_imputer(df, 'BloodPressure')
median_imputer(df, 'Insulin')
```

```python
# import the necessary libraries
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.model_selection import train_test_split
```

```python
# Perform stratified shuffle split on the data to get appropriate distribution of the outcome
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df["Outcome"]):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
```

```python
# the distribution of 1s and 0s in my strat_train_set
strat_train_set['Outcome'].value_counts() / len(strat_train_set)
```

```python
# The distribution of 1s and 0s in my full dataset
df['Outcome'].value_counts() / len(df)
```

- The distribution are roughly equal

```python
# the training features
X_train = strat_train_set.drop('Outcome', axis=1)
X_train.head(3)
```

```python
# the training labels
y_train = strat_train_set['Outcome']
y_train.head(3)
```

```python
X_test = strat_test_set.drop('Outcome', axis=1) # test features
y_test = strat_test_set['Outcome'] # test labels
```

```python
# split the data into training and validation set
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
```

```python
# import standard scaler library
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
# scale the training data
scaler.fit(X_train)
scaled_train = scaler.transform(X_train)
```

```python
# scaling the validation data
scaler.fit(X_val)
scaled_val = scaler.transform(X_val)
```

```python
# scaling the test data
scaler.fit(X_test)
scaled_test = scaler.transform(X_test)
```

## Building Models

```python
from sklearn.metrics import accuracy_score
```

```python
# Import the models 

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import SGDClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC
```

```python
# creating the models' object

log_reg = LogisticRegression(solver='liblinear')
sgd_clf = SGDClassifier()
rfc = RandomForestClassifier(n_estimators=100, max_depth=4, random_state=50)
gbrt = GradientBoostingClassifier()
dec_tree = DecisionTreeClassifier()
svc = SVC()
xgb = XGBClassifier(booster='gbtree', n_estimators=10)
```

```python
# creating the voting classifier object

voting_clf = VotingClassifier(
estimators=[('lgr', log_reg), ('rfc', rfc), ('svc', svc), ('stg', sgd_clf), 
           ('gbc',gbrt), ('dct', dec_tree), ('xgb',xgb)],
voting='hard')

voting_clf.fit(X_train, y_train)
```

```python
# Training and making predictions on the models 

from sklearn.metrics import accuracy_score
for clf in (log_reg, rfc, svc, sgd_clf, gbrt, dec_tree,xgb, voting_clf):
    clf.fit(scaled_train, y_train)
    y_pred = clf.predict(scaled_val)
    print(clf.__class__.__name__, accuracy_score(y_val, y_pred))
```

### The best performing models are: 
- `LogisticRegression`: with an accuracy of 77%
- `RandomForestClassifier` with an accuracy of 73%
- `VotingClassifier` with an accuracy of 76%

### PREDICTING THE TEST SET

#### LOGISTIC REGRESSION

```python
log_pred = log_reg.predict(scaled_test)
```

```python
print(accuracy_score(y_test, log_pred))
```

```python
rfc_pred = rfc.predict(scaled_test)
```

```python
voting_clf_pred = voting_clf.predict(scaled_test)
```
